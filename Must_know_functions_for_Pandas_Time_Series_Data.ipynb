```python
import yfinance as yf
df = yf.download("AAPL", interval="5m", period="60d")
```

# Select rows from specific daily intervals

```python
df.between_time("09:30", "09:45")["High"]
```

```python
df.at_time("09:30")
```

# Select rows based on dates

```python
df.loc["2021-09-01":]
```

# Select rows based on monthly seasonality

```python
import yfinance as yf
df = yf.download("AAPL", interval="1d", period="5Y")
```

```python
df.loc[df.index.month == 12]["High"]
```
