```python
import pandas as pd
from datetime import date,timedelta
end_date = date.today()
start_date = end_date - timedelta(days=29)
start_date_list = [start_date + timedelta(days=7*i) for i in range(5)]
end_date_list = [start_date_list[i+1] - timedelta(days=1) for i in range(4)] + [end_date]

df_1_min = pd.DataFrame()

for idx in range(5):
  data = yf.download(stock_list,start=start_date_list[idx].strftime("%Y-%m-%d"),end=end_date_list[idx].strftime("%Y-%m-%d"),interval="1m")
  df_1_min = df_1_min.append(data)
df_1_min.head(3)
```
