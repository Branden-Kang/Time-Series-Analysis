# Getting the Required Stock Historical


```python
import pandas as pd
import pandas_datareader as pdr
ts = pdr.av.time_series.AVTimeSeriesReader('TSLA',api_key='44T41340509IQGHL')
df = ts.read()
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')
df.to_csv('./output.csv')
```

# Separating Data for Training and Predictions


```python
dataset = pd.read_csv('./output.csv')
data = dataset.sample(frac=0.9)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data for Predictions: ' + str(data_unseen.shape))
```

# Creating a PyCaret Setup Instance to Start Training


```python
!pip install markupsafe==2.0.1
```

```python
from pycaret.regression import *
exp_reg102 = setup(data = data, target='close', session_id=123,use_gpu=True)
```

```python
compare_models()
```

# Creating the Model


```python
et = create_model('en')
```

```python
tuned_et = tune_model(et, n_iter=1000)
```

# Predicting Tesla Stocks

```python
unseen_predictions = predict_model(tuned_et, data=data_unseen)
unseen_predictions
```
